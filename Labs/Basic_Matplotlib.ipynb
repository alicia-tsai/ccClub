# Part 2: Simple Plots

## Line plot

- To create a empyt figure, use `plt.figure()` and then plot on your figure using `plt.plot(x, y)`.

````python
# create data
x = np.linspace(0, 10, 100)

# create an empty figure
fig = plt.figure()

# plot the figure
y = np.sin(x)
plt.plot(x, y)
````

````python
# just plot the lines, but we won't be able to save it
plt.plot(x, np.cos(x), '--', color='blue')
````

````python
# plot two lines one the same figure

plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
plt.plot(x, x+1, '--', color='r')
````

### More styles

````python
plt.plot(x, x + 0, linestyle='solid', color='red')
plt.plot(x, x + 2, linestyle='dashed', color='blue')
plt.plot(x, x + 4, linestyle='dashdot', color='green')
plt.plot(x, x + 6, linestyle='dotted', color='purple')
````

````python
# equivalent to the above
plt.plot(x, x + 0, linestyle='-', color='red')     # solid
plt.plot(x, x + 2, linestyle='--', color='blue')   # dashed
plt.plot(x, x + 4, linestyle='-.', color='green')  # dashdot
plt.plot(x, x + 6, linestyle=':', color='purple')  # dotted
````

````python
plt.plot(x, x ** 1, linestyle='solid', color='red')
plt.plot(x, x ** 2, linestyle='dashed', color='blue')
plt.plot(x, x ** 2.5, linestyle='dashdot', color='green')
plt.plot(x, x ** 3, linestyle='dotted', color='purple')
````

## Adjust plot axes

- Use `plt.xlim` and `plt.ylim` to adjust the axes of your plot.

````python
x = np.linspace(1, 10, 100)

plt.plot(x, x + 10)

plt.xlim(-1, 15)
plt.ylim(5, 25)
````

## Labeling your plot

- 4 essentail elements for a graph:
    - Axes
    - Title
    - X label
    - Y label
- If there are more lines or graphs on a figure, make sure to put the appropiate lables on them.

````python
# you can also use ':b' as a shortcut for sytles
plt.plot(x, x**3, ':b')

plt.title("Cubic polynomial")
plt.xlabel("x")
plt.ylabel("x to the power of 3")
````

### Add legend

````python
plt.plot(x, x**3, '-g', label='x to the power of 3')
plt.plot(x, x**4, ':b', label='x to the power of 4')

plt.legend()
````

## Scatter plot

````python
x = np.linspace(1, 10, 10)

plt.scatter(x, x + 10, marker='+', color='blue')
plt.scatter(x, x + 20, color='red')
plt.scatter(x, np.sin(x), color='green')
````

## Histogram

````python
x = np.random.randn(1000)

plt.hist(x)
````

### Set bins

````python
plt.hist(x, bins=50, color='steelblue')
1
````

### Plot different histograms on the same figure

- Set transparancy `alpha` to plot multiple histograms on the same figure.

````python
# use numpy's random normal function: np.random.normal(mean, variance, num)
x1 = np.random.normal(0, 1, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(2, 1, 1000)

# set alpha to adjust the tranparency
plt.hist(x1, bins=40, color='red')
plt.hist(x2, bins=40, color='green')
plt.hist(x3, bins=40, color='blue')
1
````

````python
# use numpy's random normal function: np.random.normal(mean, variance, num)
x1 = np.random.normal(0, 1, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(2, 1, 1000)

# set alpha to adjust the tranparency
plt.hist(x1, alpha=0.5, bins=40, color='red')
plt.hist(x2, alpha=0.5, bins=40, color='green')
plt.hist(x3, alpha=0.5, bins=40, color='blue')
1
````
